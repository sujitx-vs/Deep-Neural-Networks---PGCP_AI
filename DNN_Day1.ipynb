#Day 1 - Deep Neural Network.

- Basics of Tensor Flow

```python
#Imports
import tensorflow as tf
```

## TensorFlow Variables & Tensors

```python
# Single number

scalar=tf.Variable(10)
print(scalar)
scalar.assign(20)
print(scalar)
```

```python
# Vector - 1D Array
vector = tf.constant([1,5,7])
print(vector)
```

```python
# Matrix - 2D Array.
matrix = tf.constant([[1,2,3],[4,5,6]])
print(matrix)
print()

print(tf.reshape(matrix,(3,2)))
print()

print(tf.reshape(matrix,(-1)))
```

##GradientTape (Automatic Differentiation)

```python
x = tf.Variable(3.0)

with tf.GradientTape() as tape:
  y=x**2

grad = tape.gradient(y,x)
print(grad)
```

## Three Types of Summation in constant

```python
a=tf.constant([
    [1,2],
    [3,4],
    [5,6]
])
# sum everything
print(tf.reduce_sum(a))
#sum along columns
print(tf.reduce_sum(a,axis=1))
#sum along rows
print(tf.reduce_sum(a,axis=0))
```

## Activation Functions

```python
#relu
tf.keras.layers.Dense(64,activation='relu')
x=tf.constant([-2,0.0,3.0])
print(tf.nn.relu(x))

#sigmoid
tf.keras.layers.Dense(64,activation='sigmoid')
x=tf.constant([-2,0.0,3.0])
print(tf.nn.sigmoid(x))
```

```python
x = tf.constant([1,2,3])
print(tf.cast(x, tf.float32))
print(tf.cast(x,tf.bool))
```

#Adam Optimizer

```python
w = tf.Variable(0.0)

optimizer = tf.keras.optimizers.Adam(learning_rate=0.1)
```

```python
for i in range(10):
  with tf.GradientTape() as tape:
    loss=(w-4)**2
  grad = tape.gradient(loss,w)
  optimizer.apply_gradients([(grad,w)])
  print("step:",i,
        "weight:",round(w.numpy(),2),
        "loss:",round(loss.numpy(),4)
  )
```

#SGD+Momentum Optimizer

```python
w = tf.Variable(0.0)

optimizer = tf.keras.optimizers.SGD(learning_rate=0.1,momentum=0.9)
```

```python
for i in range(10):
  with tf.GradientTape() as tape:
    loss=(w-4)**2
  grad = tape.gradient(loss,w)
  optimizer.apply_gradients([(grad,w)])
  print("step:",i,
        "weight:",round(w.numpy(),2),
        "loss:",round(loss.numpy(),4)
  )
```

#RMSProp Optimizer

```python
w = tf.Variable(0.0)

optimizer = tf.keras.optimizers.RMSprop(learning_rate=0.1,rho=0.9)
```

```python
for i in range(10):
  with tf.GradientTape() as tape:
    loss=(w-4)**2
  grad = tape.gradient(loss,w)
  optimizer.apply_gradients([(grad,w)])
  print("step:",i,
        "weight:",round(w.numpy(),2),
        "loss:",round(loss.numpy(),4)
  )
```
